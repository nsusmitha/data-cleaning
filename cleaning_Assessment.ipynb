# 4.Check the content of the column- ' Author’ and define a function to clean the value. And split the name to first name and last name. (10 points)

```python
def clean_author ():
    df['first_name']=df['Author'].str.split(',',1).str[1].str.strip()
    df['last_name']=df['Author'].str.split(',',1).str[0].str.strip()
    return df
clean_author()
```

# 5.	Check the content of the column- ‘Title’ and define a function to clean the value. E.g., the original value: Walter Forbes. [A novel.] By A. A to: Walter Forbes;  Love the Avenger. By the author of “All for Gr.. to Love The Avenger. (10 points)

```python
def clean_title():
    df['Title'] = df['Title'].str.strip('[')
    df['Title'] = df['Title'].str.split('.',1).str[0].str.strip()
    return df
clean_title()
```

# 6.	Check the content of the column- ‘Place of Publication’ and define a function to clean the value. E.g., the original value: London; Virtue & Yorston to: London. (10 points)

```python
def clean_place_of_publication():
    df['Place of Publication'] = np.where(df['Place of Publication'].str.contains('London'), 'London',
                                           df['Place of Publication'].str.replace('-', ' '))
    return df
clean_place_of_publication()

```
